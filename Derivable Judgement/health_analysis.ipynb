# Part B: Data Analysis & Hypothesis Testing
This notebook performs statistical analysis on a health dataset.

# Dataset Description

The dataset contains 1200 health records including:
- Demographics (age, gender, region)
- Lifestyle factors (smoking, exercise)
- Health metrics (BMI, blood pressure, glucose, cholesterol)
- Disease indicators (diabetes, hypertension)

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

data = pd.read_csv('health_dataset.csv')
data.head()
```

# Hypothesis Formulation
H0: Smoking has no effect on diabetes
H1: Smoking affects diabetes

H0: Exercise has no effect on blood pressure
H1: Exercise affects blood pressure

# Confidence Interval

```python
mean_bp = data['blood_pressure'].mean()
std_bp = data['blood_pressure'].std()
n = len(data)

z = stats.norm.ppf(0.975)
margin_error = z * (std_bp / np.sqrt(n))

ci_lower = mean_bp - margin_error
ci_upper = mean_bp + margin_error

print(mean_bp, ci_lower, ci_upper)
```

# Significance Level

We use α = 0.05 for all statistical tests.

# T-Test Result & Interpretation

Decision Criteria:
- If p-value < 0.05 → Reject H0
- If |t-statistic| > critical value → Reject H0

```python
smokers = data[data["smoking_status"] == "Smoker"]["glucose_level"]
non_smokers = data[data["smoking_status"] == "Non-Smoker"]["glucose_level"]

t_stat, p_val = stats.ttest_ind(smokers, non_smokers)

df = len(smokers) + len(non_smokers) - 2
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)

print("T-Statistic:", t_stat)
print("P-Value:", p_val)
print("Critical Value:", t_critical)

# Final Decision
if p_val < alpha:
    print("\nDecision: Reject H0")
    print("Interpretation: Smoking has a significant effect on glucose levels.")
else:
    print("\nDecision: Accept H0")
    print("Interpretation: Smoking does not significantly affect glucose levels.")
```

# Chi-Square Test Result & Interpretation

```python
cont = pd.crosstab(data["smoking_status"], data["diabetes"])

chi2, p, dof, exp = stats.chi2_contingency(cont)

alpha = 0.05
chi_critical = stats.chi2.ppf(1 - alpha, dof)

print("Chi-Square Statistic:", chi2)
print("P-Value:", p)
print("Critical Value:", chi_critical)

# Final Decision
if p < alpha:
    print("\nDecision: Reject H0")
    print("Interpretation: Smoking and diabetes are significantly related.")
else:
    print("\nDecision: Accept H0")
    print("Interpretation: Smoking and diabetes are independent.")
```

# ANOVA Test Result & Interpretation

```python
groups = [g["blood_pressure"].values for _, g in data.groupby("age_group")]

f_stat, p_val = stats.f_oneway(*groups)

k = len(groups)
n = len(data)

df_between = k - 1
df_within = n - k

alpha = 0.05
f_critical = stats.f.ppf(1 - alpha, df_between, df_within)

print("F-Statistic:", f_stat)
print("P-Value:", p_val)
print("Critical Value:", f_critical)

# Final Decision
if p_val < alpha:
    print("\nDecision: Reject H0")
    print("Interpretation: Blood pressure differs significantly across age groups.")
else:
    print("\nDecision: Accept H0")
    print("Interpretation: No significant difference in blood pressure across age groups.")
```
